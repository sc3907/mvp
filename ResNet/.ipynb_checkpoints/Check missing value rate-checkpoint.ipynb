```python
#Check missing value rate in raw data (w/o REVEL and MPC score) and final_anno_data (w/ REVEL and MPC, -1.0 for missing value)
import csv
import matplotlib.pyplot as plt
import pandas as pd

from IPython.display import display
from matplotlib.backends.backend_pdf import PdfPages

%matplotlib inline
```

```python
fins = ['../data/case_control/case.anno.rare.csv', 
        '../data/case_control/DDD_new_0.2.anno.rare.csv',
        '../data/case_control/chd_yale.anno.rare.csv',
        '../data/case_control/case_MarkDaly.anno.rare.csv',
        '../data/case_control/control_MarkDaly.anno.rare.csv']

ret = []
names = []

    
for fin in fins:
    df = pd.read_csv(fin)
    tmp = df[~df.aaalt.isin(['.', 'X'])]
    tmp = tmp[~tmp.aaref.isin(['U'])]
    if fin == '../data/case_control/case.anno.rare.csv':
        #print tmp.head()
        tmp = tmp[tmp.disease.isin(['ASD'])]
        #print tmp.head()

    missing_rate = (tmp.astype(str) == '.').sum() / tmp.shape[0]
    names.append(fin.split('/')[-1])
    ret.append(missing_rate)
tab=pd.concat(ret, 1, keys = names)   
tab.to_csv('../data/case_control/missing_value_noREVEL_MPC.csv', sep=' ')
tab #here is table of scores except for REVEL and MPC
```

```python
fins = ['../data/case_control/case.anno.rare.HIS.reformat.cnn.csv', 
        '../data/case_control/case.anno.rare.HS.reformat.cnn.csv',
        '../data/case_control/case_MarkDaly.anno.rare.HIS.reformat.cnn.csv',
        '../data/case_control/case_MarkDaly.anno.rare.HS.reformat.cnn.csv',
        '../data/case_control/chd_yale.anno.rare.HIS.reformat.cnn.csv',
        '../data/case_control/chd_yale.anno.rare.HS.reformat.cnn.csv',
        '../data/case_control/DDD_new_0.2.anno.rare.HIS.reformat.cnn.csv',
        '../data/case_control/DDD_new_0.2.anno.rare.HS.reformat.cnn.csv',
        '../data/case_control/control_MarkDaly.anno.rare.HIS.reformat.cnn.csv',
        '../data/case_control/control_MarkDaly.anno.rare.HS.reformat.cnn.csv']

ret = []
names = []

    
for fin in fins:
    df = pd.read_csv(fin)
    tmp = df
    if fin == '../data/case_control/case.anno.rare.HIS.reformat.GCcorrected.cnn.csv':
        tmp = tmp[tmp.disease.isin(['ASD'])]
        #print tmp.head()
    if fin == '../data/case_control/case.anno.rare.HS.reformat.GCcorrected.cnn.csv':
        tmp = tmp[tmp.disease.isin(['ASD'])]
    missing_rate = (tmp.astype(str) == '-1.0').sum() / tmp.shape[0]
    names.append(fin.split('/')[-1])
    ret.append(missing_rate)
table = pd.concat(ret, 1, keys = names)
table.to_csv('../data/case_control/missing_value_withREVEL_MPC.csv')
table #here is table with REVEL and MPC computing HIS and HS separately
```
