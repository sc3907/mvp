# correlation with target in train

```python
## import csv
import matplotlib.pyplot as plt
import pandas as pd

from IPython.display import display
from matplotlib.backends.backend_pdf import PdfPages

%matplotlib inline
```

```python
df_train = pd.read_csv('../data/input_data.HS.csv')
cols = [col for col in df_train.columns if col not in {'disease', 'source', 'hg19_pos(1-based)'}]
df_train = df_train[cols]
```

```python
target = df_train.pop('target')
corr = df_train._get_numeric_data().corrwith(target)
```

```python
fig = plt.figure(figsize=(10,25))
ax = fig.add_subplot(111)
#plt.figure()
corr.plot(kind='barh', ax = ax, fontsize=15)
ax.set_xlim(-0.4,0.7)
title='correlation with target in HS training data'
ax.set_title(title,fontsize= 20) 

fig.savefig("../HS_corr.pdf", bbox_inches='tight')
```

# all mvp dist

```python
import numpy as np
pli = np.load('../data/gene/pli.npy').item()
HS_gene = set(gene for gene, pli_score in pli.iteritems() if pli_score < 0.5)
HIS_gene = set(gene for gene, pli_score in pli.iteritems() if pli_score >= 0.5)
```

```python

mvp_hs, mvp_his = [], []
fname = '/data/hq2130/large_files/All_rare_missense0624.txt'
with open(fname) as f:
    head = f.readline().strip().split()
    for line in f:
        lst = line.strip().split()
        if lst[-2] in HIS_gene:
            mvp_his.append(float(lst[-1]))
        if lst[-2] in HS_gene:
            mvp_hs.append(float(lst[-1]))
```

```python
fig = plt.figure(figsize=(8,8))




n, bins, patches = plt.hist(mvp_hs, 40)

# # add a 'best fit' line
# y = mlab.normpdf( bins, mu, sigma)
# l = plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('mvp score')
plt.ylabel('# variants')
plt.title('mvp dist among all missense variants in HS genes')
fig.savefig("mvp dist among all missense variants in HS genes.pdf", bbox_inches='tight')
```

```python
fins = ['../data/case_control/case.anno.rare.csv', 
        '../data/case_control/DDD_new_0.2.anno.rare.csv',
        '../data/case_control/chd_yale.anno.rare.csv',
        '../data/case_control/case_MarkDaly.anno.rare.csv',
        '../data/case_control/control_MarkDaly.anno.rare.csv']

ret = []
names = []

    
for fin in fins:
    df = pd.read_csv(fin)
    tmp = df[~df.aaalt.isin(['.', 'X'])]
    tmp = tmp[~tmp.aaref.isin(['U'])]
    if fin == '../data/case_control/case.anno.rare.csv':
        #print tmp.head()
        tmp = tmp[tmp.disease.isin(['ASD'])]
        #print tmp.head()

    missing_rate = (tmp.astype(str) == '.').sum() / tmp.shape[0]
    names.append(fin.split('/')[-1])
    ret.append(missing_rate)
tab=pd.concat(ret, 1, keys = names)   
tab.to_csv('../data/case_control/missing_value_noREVEL_MPC.csv', sep=' ')
tab #here is table of scores except for REVEL and MPC
```
